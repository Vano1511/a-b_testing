## мой продукт: онлайн игра Lords mobile


### 1 метрика: DAU(dayli active users) уникальные активные пользователи в день
##### Гипотеза: если ввести бонусы за ежедневное посещение игры, то DAU вырастет на 20%.

### 2 метрика: Конверсия в покупку (процент зарегестрированных игроков, которые осуществили покупку продукта)
##### Гипотеза: если ввести бонус за первую покупку (донатный герой в подарок), то конверсия увеличится на 15%

### 3 метрика: Бонус за возвращение неактивных пользователей.
##### Гипотеза: если ввести бонус за возврат игрока в игру (пакет ускорений и щитов) то будет возвращаться на 10% бывших игроков больше.

### 4 метрика: увеличение среднего месячного чека с игрока
##### Гипотеза: если ввести бонус (два в цене одного), то средний месячный чек с платящего пользователя вырастет на 7%

### 5 Метрика: Оптимизация серверного пространства, приходящегося на продукт.
##### Гипотеза: если ввести минимальный порог активных игроков в старых (болше 6 месяцев) королевствах и предоставления им бесплатного свитка миграции и срока недели на миграцию, а потом закрытие королевства, то затраты на сервера уменьшатся на 10%, в то время, как отток игроков (в основном бесплатных пользователей) увеличится на 2%


### Приоритезация метрик (ICE) влияние-уверенность-легкость реализации

## Шаблон первого эксперимента

#### 1. Гипотеза: если ввести бонус (два в цене одного), то средний месячный чек с платящего пользователя вырастет на 7%
#### 2. Что делаем : вводим бонус с периодичностью два раза в месяц
#### 3. На ком тестируем: на тех, ползователей, кто сделал хоть одну покупку в игре (средний месячный чек - 22,5 доллара)  
#### 4. Ключевые метрики: средний месячный чек с платящего пользователя 
1. так как игра популярная и пользователей в ней много (более 500 тыс. в день) 
2. и конверсия в платящего тоже высокая (14.3%)
3. можем повысить стандартное отклонение до 100
4. можем повысить мощность теста до 99%
5. снизим вероятность ошибки до 1%
6. колькулятор рассчета выборки выдал 156216 пользователей, что можно проверить за несколько дней
7. так как у нас метрика месячная, - планируем провести испытание в течении двух месяцев(чтобы усреднить, если будут выбросы)

#### 5. Ожидается увеличение среднего месяцного чека с платящего до 24.1 доллара
#### 6. План дейсвтий:
- отвергаются всякие преждевременные выводы до окончания срока теста (даже если тест был прерван);
- если ключевая метрика поднимется выше уровня в 23,5 доллара, то принимаем гипотезу и выводим изменение на всех пользователей;
- в остальных случаях - откатываем изменения;

## Шаблон второго эксперимента

#### 1. Гипотеза: если ввести бонусы за ежедневное посещение игры, то DAU вырастет на 20%.
#### 2. Что делаем : вводим бонусы с нарастанием их количества в зависимости от посещений подряд на протяжении 21 дня
#### 3. На ком тестируем: на всех пользователях 
#### 4. Ключевые метрики: Daily Active Users (DAU) на данный момент - 58,6%
1. так как игра популярная и пользователей в ней много (более 500 тыс. в день) 
2. можем повысить стандартное отклонение до 100
3. можем повысить мощность теста до 95%
4. снизим вероятность ошибки до 1%
5. колькулятор рассчета выборки выдал очень смешную цифру, что можно проверить за несколько минут
6. так как у нас метрика дневная, - планируем провести испытание в течении трех недель(чтобы усреднить, если будут выбросы)

#### 5. Ожидается увеличение метрики до 70%
#### 6. План дейсвтий:
- отвергаются всякие преждевременные выводы до окончания срока теста (даже если тест был прерван);
- если ключевая метрика поднимется выше уровня в 60%, то принимаем гипотезу и выводим изменение на постоянной основе;
- в остальных случаях - откатываем изменения;
