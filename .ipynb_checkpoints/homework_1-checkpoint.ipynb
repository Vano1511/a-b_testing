## мой продукт: онлайн игра Lords mobile


### 1 метрика: DAU(dayli active users) уникальные активные пользователи в день
##### Гипотеза: если ввести бонусы за ежедневное посещение игры, то DAU вырастет на 20%.

### 2 метрика: Конверсия в покупку (процент зарегестрированных игроков, которые осуществили покупку продукта)
##### Гипотеза: если ввести бонус за первую покупку (донатный герой в подарок), то конверсия увеличится на 15%

### 3 метрика: Бонус за возвращение неактивных пользователей.
##### Гипотеза: если ввести бонус за возврат игрока в игру (пакет ускорений и щитов) то будет возвращаться на 10% бывших игроков больше.

### 4 метрика: увеличение среднего месячного чека с игрока
##### Гипотеза: если ввести бонус (два в цене одного), то средний месячный чек с платящего пользователя вырастет на 7%

### 5 Метрика: Оптимизация серверного пространства, приходящегося на продукт.
##### Гипотеза: если ввести минимальный порог активных игроков в старых (болше 6 месяцев) королевствах и предоставления им бесплатного свитка миграции и срока недели на миграцию, а потом закрытие королевства, то затраты на сервера уменьшатся на 10%, в то время, как отток игроков (в основном бесплатных пользователей) увеличится на 2%
